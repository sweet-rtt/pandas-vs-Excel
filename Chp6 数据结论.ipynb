# 3.线性回归，数据预测

```python
import pandas as pd
import matplotlib.pyplot as plt

sales = pd.read_excel('sales.xlsx', dtype={'month':str})
print(sales)
```

```python
plt.bar(sales.index, sales.revenue)
plt.title('sales')
plt.xticks(sales.index, sales.month, rotation=90)
# 紧凑布局 plt.tight_layout()
plt.show()
```

```python
from scipy.stats import linregress

slope, intercept, r, p, std_err = linregress(sales.index, sales.revenue)
exp = sales.index * slope + intercept

plt.scatter(sales.index, sales.revenue)
plt.plot(sales.index, exp, color='orange')
plt.title(f'y = {slope} * x + {intercept}')
plt.xticks(sales.index, sales.month, rotation=90)
# 紧凑布局 plt.tight_layout()
plt.show()
```

```python
# 预测2019年12月的revenue
print(slope*35+intercept)
```

# 4.条件格式（上）

```python
import pandas as pd

def low_score_red(s):
    color = 'red' if s<60 else 'black'
    return f'color:{color}'

def highest_score_green(col):
    return ['background-color:lime' if s==col.max() else 'background-color:white' for s in col]

students = pd.read_excel('./students3.xlsx')
students.style.applymap(low_score_red, subset=['test_1', 'test_2', 'test_3']) \
.apply(highest_score_green, subset=['test_1', 'test_2', 'test_3'])
```

# 6.行操作集锦-移花接木

```python
import pandas as pd

page_001 = pd.read_excel('./students4.xlsx', sheet_name='page_001')
page_002 = pd.read_excel('./students4.xlsx', sheet_name='page_002')

# 1.连接两张表，sql里的unit操作
student = page_001.append(page_002).reset_index(drop=True)
print(student)
```

```python
# 2.末尾追加手动创建的一行
stu = pd.Series({'id':41, 'name':'Abel', 'score':99})
student = student.append(stu, ignore_index=True)
print(student)
```

```python
# 3.更改数据表里的值
student.at[39, 'name'] = 'Beily'
student.at[39, 'score'] = 120
print(student)

# 方法2，直接替换一行，上一种方法未替换
stu2 = pd.Series({'id':41, 'name':'Beily', 'score':120})
student.iloc[39] = stu2
print(student)
```

```python
# 4.在数据表中插入一行
stu3 = pd.Series({'id':101, 'name':'Danni', 'score':101})
part1 = student[:20]
part2 = student[20:]
student = part1.append(stu3, ignore_index=True).append(part2).reset_index(drop=True)
print(student)
```

```python
# 5.删除数据行，删除index为0,1,2的行
student.drop(index=[0, 1, 2], inplace=True)
print(student)
# 方法2
student.drop(index=range(0, 3), inplace=True)
print(student)
# 方法3：用切片删除
student.drop(index=student[0:3].index, inplace=True)
```

```python
# 6.有条件的删除
for i in range(5, 15):
    student['name'].at[i] = ''

missing = student.loc(student['name'] == '')
student.drop(index=missing.index, inplace=True)
student = student.reset_index(drop=True)
print(student)
```

# 7.列操作集锦-搬砖侠

```python
import pandas as pd

page_001 = pd.read_excel('./students4.xlsx', sheet_name='page_001')
page_002 = pd.read_excel('./students4.xlsx', sheet_name='page_002')

# 1.连接，concat功能与append类似：1.append（2）， pd.concat（[1, 2]）
student = pd.concat([page_001, page_002], axis=1).reset_index(drop=True)
print(student)
```

```python
# 2.追加一列
student['age']=25
# 或者 student['age'] = np.repeat(25, len(student))
# 序列 student['age'] = np.arange(0, len(student))

print(student)
```

```python
# 3.删除某列
student.drop(columns=['age', 'score'], inplace=True)
print(student)
```

```python
# 4.在id和name之间插入一列foo列
import numpy as np
student.insert(1, column='foo', value=np.repeat('foo', len(student)))
print(student)
```

```python
# 5.将列名改为大写
student.rename(columns={'foo':'FOO', 'name':'NAME'}, inplace=True)

print(student)
```

```python
# 6.删去空值
student['id'] = student['id'].astype(float)

for i in range(5, 15):
    student['id'].at[i] = np.nan

student.dropna(inplace=True)

print(student)
```
