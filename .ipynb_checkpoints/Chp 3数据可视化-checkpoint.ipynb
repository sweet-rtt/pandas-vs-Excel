# 1.柱状图

```python
import pandas as pd
import matplotlib.pyplot as plt

subjects = pd.read_excel('./subjects.xlsx')
print(subjects)
subjects.sort_values(by='number', inplace=True, ascending=False)
# 1.常规pandas画图
subjects.plot.bar(x='field', y='number', color='orange', title='International students by field')
# 紧凑型布局
# plt.tight_layout()
plt.show()
```

```python
# 1.matplotlib画图
plt.bar(subjects.field, subjects.number)
# x轴标签重叠，故把标签旋转
plt.xticks(subjects.field, rotation=90)
# 加x轴名称，y轴名称以及图标名称
plt.xlabel('field')
plt.ylabel('number')
plt.title('International students by fields', fontsize=16)
# 紧凑型布局
# plt.tight_layout()
plt.show()
```

# 2.分组柱图，深度优化

```python
import pandas as pd
import matplotlib.pyplot as plt

subjects2 = pd.read_excel('./subjects2.xlsx')
subjects2.sort_values(by=2017, inplace=True, ascending=False)
subjects2.plot.bar(x='field', y=[2016, 2017], color=['orange', 'red'], title='International students by field')
# 给x轴y轴加标签名
plt.xlabel('field', fontweight='bold')
plt.ylabel('number', fontweight='bold')
plt.title('International students by field', fontsize=16, fontweight='bold')
# x轴标签旋转,gca拿到轴，gcf拿到figures
ax = plt.gca()
ax.set_xticklabels(subjects2.field, rotation=45, ha='right')
# 调整figures左方和下方的间距
f = plt.gcf()
f.subplots_adjust(left=0.2, bottom=0.4)
plt.show()
```

```python
import numpy as np
plt.bar(subjects2.field, subjects2[2016], width=0.35)
plt.bar(subjects2.field+'0.35', subjects2[2017], width=0.35)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 数据
N = 5
x = [20, 10, 30, 25, 15]
y = np.arange(N)

# 绘图 x= 起始位置， bottom= 水平条的底部(左侧), y轴， height 水平条的宽度， width 水平条的长度
p1 = plt.bar(x=0, bottom=y, height=0.5, width=x, orientation="horizontal")

# 展示图形
plt.show()
```

# 3.叠加柱状图，水平柱状图

```python
import pandas as pd
import matplotlib.pyplot as plt

users = pd.read_excel('./users.xlsx')
users['total'] = users['Oct'] + users['Nov'] + users['Dec']
users.sort_values(by='total', inplace=True, ascending=False)
users.plot.bar(x='Name', y=['Oct', 'Nov', 'Dec'], stacked=True, title='User Behavior')
plt.show()
```

```python
users.sort_values(by='total', inplace=True)
users.plot.barh(x='Name', y=['Oct', 'Nov', 'Dec'], stacked=True, title='User Behavior')
plt.show()
```

# 4.饼图

```python
import pandas as pd
import matplotlib.pyplot as plt

student = pd.read_excel('./student.xlsx', index_col='from')
student['2017'].sort_values(ascending=True).plot.pie(fontsize=7, startangle=-270)
plt.title('Source of International Students', fontsize=16, fontweight='bold')
plt.ylabel('2017', fontsize=16, fontweight='bold')
plt.show()
```

```python
student['2017'].plot.pie(fontsize=7)
plt.title('Source of International Students', fontsize=16, fontweight='bold')
plt.ylabel('2017', fontsize=16, fontweight='bold')
plt.show()
```

```python
student['2017'].plot.pie(fontsize=7, counterclock=False)
plt.title('Source of International Students', fontsize=16, fontweight='bold')
plt.ylabel('2017', fontsize=16, fontweight='bold')
plt.show()
```

# 5.折线趋势图，叠加区域图

```python
import pandas as pd
import matplotlib.pyplot as plt

weeks = pd.read_excel('./weeks.xlsx', index_col='week')
print(weeks)
print(weeks.columns)
weeks.plot(y=['accessories', 'bikes', 'clothing', 'components'])
plt.title('Sales Weekly Trend', fontsize=16, fontweight='bold')
plt.ylabel('Total', fontsize=16, fontweight='bold')
plt.xticks(weeks.index, fontsize=8, fontweight='bold')
plt.show()
```

```python
# 叠加区域图
weeks.plot.area(y=['accessories', 'bikes', 'clothing', 'components'])
plt.title('Sales Weekly Trend', fontsize=16, fontweight='bold')
plt.ylabel('Total', fontsize=16, fontweight='bold')
plt.xticks(weeks.index, fontsize=8, fontweight='bold')
plt.show()
```

```python
# 叠加柱状图
weeks.plot.bar(y=['accessories', 'bikes', 'clothing', 'components'], stacked=True)
plt.title('Sales Weekly Trend', fontsize=16, fontweight='bold')
plt.ylabel('Total', fontsize=16, fontweight='bold')
plt.xticks(weeks.index, fontsize=8, fontweight='bold')
plt.show()
```

# 6.散点图、直方图

```python
import pandas as pd
import matplotlib.pyplot as plt

# 散点图
pd.options.display.max_columns=777
homes = pd.read_excel('./home_data.xlsx')
homes.head()

homes.plot.scatter(x='sqft_living', y='price')
plt.show()
```

```python
# 面积直方图
homes['sqft_living'].plot.hist(bins=100)
plt.xticks(range(0, max(homes['sqft_living']),500), fontsize=8, rotation=90)
plt.show()
```

```python
# 房价直方图
homes['price'].plot.hist(bins=100)
plt.xticks(range(0, max(homes['price']),100000), fontsize=8, rotation=90)
plt.show()
```

# 7.密度图，数据相关性

```python
# 面积密度图
homes['sqft_living'].plot.kde()
plt.xticks(range(0, max(homes['sqft_living']),500), fontsize=8, rotation=90)
plt.show()
```

```python
# 相关性
print(homes.corr())
```
